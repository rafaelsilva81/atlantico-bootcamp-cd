# Visualização #03 (Gorjetas)

```python
#Imports necessários
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

#Carrega o dataframe
df = sns.load_dataset("tips")
df.head()

#Usa o grid
sns.set_style("whitegrid")

#Define um tamanho menor para a figura
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x="day", y="total_bill", data=df, color='cornflowerblue') #O campo x é o dia da semana, o y é a conta total (em dolar)

#Altera as labels, mostra o gráfico
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
