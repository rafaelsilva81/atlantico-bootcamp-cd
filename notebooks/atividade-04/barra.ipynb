# Visualização #01 (Arrecadação de Filmes)

```python
# Importação das bibliotecas necessárias
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

# Realizando leitura de um csv a partir de um link
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')

#Divide os valores por 1 Milhão para auxiliar a visualização
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000) 

#Ordena pela coluna "arrecadação", em ordem decrescente
df = df.sort_values(by='arrecadacao', ascending=False) #Para ordernar por ordem crescente utilizar 'ascending=True'
df.head()

#Indica ao (sns) para utilizar o estilo de linhas verticais
sns.set_style("whitegrid")

plt.figure(figsize=(5, 2)) #Isso apenas reduz o tamanho da figura
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='#a029cf')

ax.bar_label(
    ax.containers[0],
    fmt='%.1f', #1 casa decimal
    label_type='center', #Centraliza a informação na barra
    color='white'
)

plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')

ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```
